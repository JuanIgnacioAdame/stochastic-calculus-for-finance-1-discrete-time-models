# Problem

A wealthy investor provides a small amount of money $X_0$ for you to use to prove the effectiveness of your investment scheme over the next $N$ periods.  You are permitted to invest in the $N$-period binomial model, subject to the condition that the value of your portfolio is never allowed to be negative.  If at time $N$ the value of your portfolio $X_N$ is at least $\gamma$, a positive constant specified the investor, then you will be given a large amount of money to manage for her.  Therefore, your problem is the following:
\begin{equation}
\text{Maximize} \quad \mathbb{P}(X_N\geq \gamma),
\end{equation}
where $X_N$ is generated by a portfolio process beginning with the initial wealth $X_0$ and where the value $X_n$ of your portfolio process satisfies
\begin{equation}
X_n\geq 0, \quad n=1,2,\dots,N.
\end{equation}

In the way that Problem 3.3.1 was reformulated as Problem 3.3.3, this problem may be reformulated as
\begin{equation}
\text{Maximize} \quad \mathbb{P}(X_N\geq\gamma)
\end{equation}
subject to
\begin{align}
&\tilde{\mathbb{E}}\frac{X_N}{(1+r)^N}=X_0,\\
&X_n\geq 0, \quad n=1,2,\dots,N.
\end{align}

(i) Show that if $X_N\geq 0$, then $X_n\geq 0$ for all $n$.

(ii) Consider the function
\begin{align}
U(x)&=0, \quad \text{ if } 0\leq x<\gamma,\\
&=1, \quad \text{ if } x\geq\gamma.
\end{align}
Show that for each fixed $y>0$, we have
\begin{equation}
U(x)-yx\leq U(I(y))-yI(y) \quad\forall x\geq 0,
\end{equation}
where
\begin{align}
I(y)&=\gamma, \quad \text{ if } 0<y\leq\frac{1}{y},\\
&=0, \quad \text{ if } y>\frac{1}{y}.
\end{align}

(iii) Assume there is a solution $\lambda$ to the equation
\begin{equation}
\mathbb{E}\left[\frac{Z}{(1+r)^N}I\left(\frac{\lambda Z}{(1+r)^N}\right)\right]=X_0.
\end{equation}
following the argument of Exercise 3.8, show that the optimal $X_N$ is given by
\begin{equation}
X_N^*=I\left(\frac{\lambda Z}{(1+r)^N}\right).
\end{equation}

(iv) As we did to obtain Problem 3.3.5, let us list the $M=2^N$ possible coin toss sequences, labeling them $\omega^1,\dots,\omega^M$, and then define $\zeta_m=\zeta(\omega^m)$, $p_m=\mathbb{P}(\omega^m)$.  However, here we list these sequences in ascending order of $\zeta_m$, i.e., we label the coin toss sequences so that
\begin{equation}
\zeta_1\leq\zeta_2\leq\dots\leq\zeta_M.
\end{equation}
Show that the assumption that there is a solution $\lambda$ to (3.6.4) is equivalent to assuming that for some positive integer $K$ we have $\zeta_K<\zeta_{K+1}$ and
\begin{equation}
\sum_{m=1}^{K}\zeta_mp_m=\frac{X_0}{\gamma}.
\end{equation}

(v) Show that $X_N^*$ is given by
\begin{align}
X_N(\omega^m)&=\gamma, \quad \text{ if } m\leq K,\\
&=0, \quad \text{ if } m\geq K+1.
\end{align}

# Answer

(i) Recall the risk-neutral pricing formula for the portfolio wealth process
\begin{equation}
X_n=\tilde{\mathbb{E}}\left[\frac{X_N}{(1+r)^{N-n}}\right].
\end{equation}
If $X_N\geq 0$, then the expectation on the right is greater than or equal to $0$, and hence $X_n\geq 0$ as well. $\square$

(ii) Our approach is to find an upper bound for the expression $U(x)-yx$, and at the same time a lower bound for $U(I(y))-yI(y)$. There are four cases, which we list below
\begin{align}
0\leq &x<\gamma \quad \text{ and } &0<y\leq\frac{1}{\gamma},\\
0\leq &x<\gamma \quad \text{ and } &y>\frac{1}{\gamma},\\
&x\geq\gamma \quad \text{ and } &0<y\leq\frac{1}{\gamma},\\
&x\geq\gamma \quad \text{ and } &y>\frac{1}{\gamma}.
\end{align}

For the first case, we have
\begin{equation}
U(x)-yx=0-yx\leq 0,
\end{equation}
and
\begin{equation}
U(I(y))-yI(y)=U(\gamma)-y\gamma=1-y\gamma\geq 1-\frac{1}{\gamma}\gamma=0\geq U(x).
\end{equation}

For the second case, we have
\begin{equation}
U(x)-yx=0-yx\leq 0,
\end{equation}
and
\begin{equation}
U(I(y))-yI(y)=U(0)-y\times 0=0-y\times 0=0\geq U(x).
\end{equation}

For the third case, we have
\begin{equation}
U(x)-yx=1-yx\leq 1-y\gamma,
\end{equation}
and
\begin{equation}
U(I(y))-I(y)=U(\gamma)-y\gamma=1-y\gamma\geq U(x).
\end{equation}

For the fourth case, we have
\begin{equation}
U(x)-yx=1-yx\leq 1-\frac{1}{\gamma}\gamma=1-1=0,
\end{equation}
\begin{equation}
U(I(y))-yI(y)=U(0)-y\times 0=0-y\times 0=0\geq U(x). \square
\end{equation}

(iii) In the part (ii) of this exercise, we showed that the function of $x$ given by $U(x)-yx$ is maximized by $x=I(y)$.  In other words,
\begin{equation}
U(x)-yx\leq U(I(y))-yI(y).
\end{equation}
Combined with the assumption that there exists a $\lambda$ such that
\begin{equation}
\mathbb{E}\left[\frac{Z}{(1+r)^N}I\left(\frac{\lambda Z}{(1+r)^N}\right)\right]=X_0,
\end{equation}
we have satisfied the hypotheses of Exercise 3.8.  Hence, the result of Exercise 3.8 (ii) immediately gives us that
\begin{equation}
X^*_N=I\left(\frac{\lambda Z}{(1+r)^N}\right).\square
\end{equation}

(iv) $\Rightarrow)$ Assume
\begin{equation}
\mathbb{E}\left[\frac{Z}{(1+r)^N}I\left(\frac{\lambda Z}{(1+r)^N}\right)\right]=X_0,
\end{equation}
or, equivalently,
\begin{equation}
\mathbb{E}\left[\zeta I\left(\lambda\zeta\right)\right]=X_0.
\end{equation}
Expanding the left hand side, we obtain
\begin{equation}
\sum_{m=1}^{M}p_m\zeta_mI(\lambda\zeta_m)=
\sum_{\substack{m\\\lambda\zeta_m>\frac{1}{\gamma}}}p_m\zeta_mI(\lambda\zeta_m)+\sum_{\substack{m\\0<\lambda\zeta_m\leq\frac{1}{\gamma}}}p_m\zeta_mI(\lambda\zeta_m).
\end{equation}

Because $I(y)=0$ whenever $y>\frac{1}{\gamma}$, the left term above drops out and we are left with
\begin{equation}
\sum_{\substack{m\\0<\lambda\zeta_m\leq\frac{1}{\gamma}}}p_m\zeta_m\gamma.
\end{equation}
Note that the set of values of $0<\lambda\zeta_m\leq\frac{1}{\gamma}$ cannot be empty, otherwise $X_0=0$, and this violates our hypothesis that $X_0>0$.
Next, we divde both sides by $\lambda$ and obtain
\begin{equation}
\sum_{\substack{m\\0<\lambda\zeta_m\leq\frac{1}{\gamma}}}p_m\zeta_m=\frac{X_0}{\gamma}.
\end{equation}

Now note that we cannot have $m$ run though all the indices (i.e. $m=1,\dots,M$), because then left-hand side of the expression would equal $\mathbb{E}\left[\zeta\right]=1$, and we'd have $1=\frac{X_0}{\gamma}$.  But $X_0$ and $\gamma$ are allowed to vary freely, and that condition would violate that.  Additionally, note that the problem is trivial if $X_0=\gamma$, because in that case the investor can guarantee that he hits the goal (of having wealth greater than or equal to $\gamma$ at time $N$) by investing in the money market at every time period.  In short, the problem is only non-trivial if $X_0<\gamma$, and this implies that $m$ cannot run through all the indices.

Now note that if $m$ cannot run through all the indices, then this is equivalent to saying that there is a positive integer $K$ such that
\begin{equation}
\sum_{m=1}^K\zeta_mp_m=\frac{X_0}{\gamma}.
\end{equation}
It only remains to show that $\zeta_K<\zeta_{K+1}$.  But this is guaranteed by the fact that, according to the way we split the summation, we would have $\lambda\zeta_{K+1}>\frac{1}{\gamma}$, and $0\leq\lambda\zeta_{K+1}\leq\frac{1}{\gamma}$, and this implies $\zeta_K<\zeta_{K+1}$.

$\Leftarrow)$ Now assume that for some positive integer $K$ we have $\zeta_{K}<\zeta_{K+1}$ and
\begin{equation}
\sum_{m=1}^{K}\zeta_mp_m=\frac{X_0}{\gamma}.
\end{equation}
This implies that
\begin{equation}
\sum_{m=1}^{K}\zeta_mp_m\gamma=X_0.
\end{equation}
Note that the above expression is equivalent to
\begin{equation}
\sum_{m=1}^{K}\zeta_mp_mI(\lambda\zeta_m)+\sum_{m=K+1}^{M}\zeta_mp_mI(\lambda\zeta_m)=X_0,
\end{equation}
if we can find a $\lambda$ such that
\begin{align}
&I(\lambda\zeta_m)=\gamma \quad \text{for } 1\leq m\leq K,\\
&I(\lambda\zeta_m)=0 \quad \text{for } K+1\leq m\leq M,
\end{align}
which is equivalent, by definition of $I(y)$, to requiring that
\begin{align}
0<&\lambda\zeta_m\leq\frac{1}{\gamma} \quad \text{for } 1\leq m\leq K,\\
&\lambda\zeta_m>\frac{1}{\gamma} \quad \text{for } K+1\leq m\leq M.
\end{align}

Because the $\zeta_m$ are indexed in ascending order, we can satisfy the two equations above if we can find a $\lambda>0$ such that
\begin{gather}
\lambda\zeta_K\leq\frac{1}{\gamma}\Rightarrow\lambda\leq\frac{1}{\gamma\zeta_K},\\
\lambda\zeta_{K+1}>\frac{1}{\gamma}\Rightarrow\lambda>\frac{1}{\gamma\zeta_{K+1}}.
\end{gather}
We can indeed find such a $\lambda$ because $\zeta_K<\zeta_{K+1}$.

In short, for appropriate $\lambda$, the expression
\begin{equation}
\sum_{m=1}^{K}\zeta_mp_mI(\lambda\zeta_m)+\sum_{m=K+1}^{M}\zeta_mp_mI(\lambda\zeta_m)=X_0
\end{equation}
is justified.  Now it is just a matter of rewriting the left hand of the expression, and obtained the desired result:
\begin{align}
\sum_{m=1}^{K}\zeta_mp_mI(\lambda\zeta_m)&+\sum_{m=K+1}^{M}\zeta_mp_mI(\lambda\zeta_m)=\sum_{m=1}^{M}\zeta_mp_mI(\lambda\zeta_m)\\
&=\mathbb{E}\left[\zeta I\left(\lambda\zeta\right)\right]\\
&=\mathbb{E}\left[\frac{Z}{(1+r)^N}I\left(\frac{\lambda Z}{(1+r)^N}\right)\right]=X_0.\square
\end{align}

(v) Assume there is a $\lambda$ satisfying
\begin{equation}
\mathbb{E}\left[\frac{Z}{(1+r)^N}I\left(\frac{\lambda Z}{(1+r)^N}\right)\right]=X_0.
\end{equation}
Now let 
\begin{equation}
X^*_N=I\left(\lambda\frac{Z}{(1+r)^N}\right)=I\left(\lambda\zeta\right).
\end{equation}
In part (iv) of this exercise, we showed that that $\lambda$ might be, it has to satisfy the conditions
\begin{align}
&I(\lambda\zeta_m)=I(\lambda\zeta(\omega^m))=\gamma \quad \text{for } 1\leq m\leq K,\\
&I(\lambda\zeta_m)=I(\lambda\zeta(\omega^m))=0 \quad \text{for } K+1\leq m\leq M.
\end{align}
Hence
\begin{align}
&X^*_N(\omega^m)=I(\lambda\zeta_m)=I(\lambda\zeta(\omega^m))=\gamma \quad \text{for } 1\leq m\leq K,\\
&X^*_N(\omega^m)=I(\lambda\zeta_m)=I(\lambda\zeta(\omega^m))=0 \quad \text{for } K+1\leq m\leq M. \square
\end{align}
